```python
import pandas as pd
import numpy as np
```

### Carregando os arquivos

```python
#importando a primeira base (Netflix)
df_netflix = pd.read_csv('netflix_titles.csv')
```

```python
#verificando a quantidade de colunas
df_netflix.shape
```

```python
#importando a segunda base (Netflix)
df_amazon = pd.read_csv('amazon_prime_titles.csv')
```

```python
#verificando a quantidade de colunas
df_amazon.shape
```

### Mesclando os modelos

```python
#merge
df = pd.merge(df_netflix, df_amazon, on='show_id', how='outer', suffixes=('_df_netflix', '_df_amazon'))
```

```python
#analisando a quantidade de linhas e colunas após a mesclagem
df.shape
```

```python
#aumentando a visuzalização de todas as colunas no novo dataframe
pd.set_option('Display.max_columns',23)
```

```python
#Criando bases iguais para analises diferentes
df_2 = df
df_3 = df
```

### Filtrando os dados

#### 1- Top 10 atores/atrizes considerando todos os dados

```python
#aumentando o tamanho de linhas
pd.set_option('display.max_colwidth', 1000)
```

```python
#'splitando' a coluna de atores e atrizes
df_2['cast_df_netflix'] = df_2['cast_df_netflix'].str.split(', ')
df_2['cast_df_amazon'] = df_2['cast_df_amazon'].str.split(', ')
df_2 = df.explode('cast_df_netflix')
df_2 = df.explode('cast_df_amazon')
```

```python
top10_atores = pd.concat([pd.Series(top_netflix_atores), pd.Series(top_amazon_atores)], axis=1)
top10_atores.head(10)
```

#### 6- A frequência de "TV Show" de todos os dados e comparativamente em relação as duas plataformas

```python
df.head(5)
```

```python
#Frequencia de todos como Netflix
tv_show_netflix = np.sum(df['type_df_netflix'] == 'TV Show')
print(f'A quantidade de frequência da Netflix como Tv Show é: {tv_show_netflix}')
```

```python
#Frequencia de todos como Amazon
tv_show_amazon = np.sum(df['type_df_amazon'] == 'TV Show')
print(f'A quantidade de frequência da Amazon como Tv Show é: {tv_show_amazon}')
```

```python
#Frequencia de todos combinados
tv_show_combinados = np.sum((df['type_df_netflix'] == 'TV Show') | (df['type_df_amazon'] == 'TV Show'))
print(f'A quantidade de frequência de todos os dados combinados como Tv Show é: {tv_show_combinados}')
```

```python
#Frequencia de todos os dados somados
tv_show_somados = tv_show_netflix + tv_show_amazon
print(f'A quantidade de frequência de todos os dados somados como Tv Show é: {tv_show_somados}')
```
